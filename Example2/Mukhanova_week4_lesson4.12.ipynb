```python
import pandas as pd
```

```python
#загружаем датафреймы
invoices01 = pd.read_excel("Invoices_part1.xlsx")
invoices02 = pd.read_excel("Invoices_part2.xlsx")
```

```python
invoices01.head()
```

```python
invoices02.head()
```

```python
#.concat()
invoices = pd.concat(objs = [invoices01,invoices02])
```

```python
#чтобы различать источник, можно добавить как key
invoices = pd.concat(objs = [invoices01,invoices02], keys = ["01","02"])
```

```python
invoices
```

```python
#.value_counts()
invoices.index.value_counts()
```

```python
#.loc сентябрь
invoices.loc[("01")]
```

```python
# параметр ingore_index=True чтобы починить индексы
invoices = pd.concat(objs = [invoices01,invoices02], ignore_index = True)
```

```python
#загружаем датафрейм
customers = pd.read_csv("Customers.csv",sep = ';')
customers.sort_index(inplace=True)
customers.head()
```

```python
customers.describe()
```

```python
customers.info()
```

```python
#inner join
invoices.merge(customers, how ="inner", on = "CustomerId").count()
```

```python
#left join
invoices.merge(customers, how ="left", on = "CustomerId").count()
```

```python
#right join
invoices.merge(customers, how ="right", on = "CustomerId").count()
```

```python
#outer join
invoices.merge(customers, how ="outer", on = "CustomerId").count()
```
