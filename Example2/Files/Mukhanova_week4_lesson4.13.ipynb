```python
import pandas as pd
import glob
```

```python
files = glob.glob("*.csv")
files
```

```python
df = pd.DataFrame()

for file in files:
    data = pd.read_csv(file)
    data["file_name"] = file
    df = pd.concat([df, data])
```
