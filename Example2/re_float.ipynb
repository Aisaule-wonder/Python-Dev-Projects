Импортируйте две библиотеки: **random** и **pandas**. </br>**random** будет использоваться для генерации случайных чисел и выбора типа ошибки формата числа

```python
import random
import pandas as pd
```

```python
df = pd.read_excel('Customers.xlsx')
df
```

```python
# Функция для генерации датасета
def generate_dataset(num_samples):
    dataset = []  # Создаем пустой список для датасета
    for _ in range(num_samples):
        # Случайным образом выбираем тип ошибки формата числа
        error_type = random.choice(['missing_decimal', 'extra_decimal', 'wrong_delimiter'])
        
        if error_type == 'missing_decimal':
            # Генерируем число без десятичной части
            number = random.randint(1, 100)
            dataset.append(str(number))
            
        elif error_type == 'extra_decimal':
            # Генерируем число с дополнительной десятичной точкой
            number = random.uniform(1, 100)
            dataset.append(str(number).replace('.', '..'))
            
        elif error_type == 'wrong_delimiter':
            # Генерируем число с неправильным разделителем десятичной части
            number = random.uniform(1, 100)
            dataset.append(str(number).replace('.', ','))
    
    return dataset

# Генерируем датасет с 10 примерами
dataset = generate_dataset(10)

# Создаем DataFrame из датасета
df = pd.DataFrame(dataset, columns=['Value'])

# Сохраняем датасет в CSV-файл
df.to_csv('data.csv', index=False)
print("Датасет сохранен в data.csv.")
```

В Python тип данных **"object"** относится к общему классу объектов, который является базовым для всех остальных типов данных.

Когда переменная имеет тип "object", это означает, что ее значение может быть любым объектом в Python. Объекты могут быть экземплярами различных классов, таких как строки (str), списки (list), словари (dict), функции (function) и т. д.

Тип "object" является динамическим типом данных, что означает, что переменная этого типа может принимать значения разных типов в разное время выполнения программы. Таким образом, переменная типа "object" может содержать данные любого типа в Python.

**3 способ с использованием библиотеки re**
 </br>
В этом примере мы используем регулярное выражение r"[-+]?\d*\.\d+|\d+", чтобы найти числа с плавающей точкой или целые числа в каждом значении столбца "Value". Затем мы преобразуем каждое найденное совпадение в тип float с помощью функции float и сохраняем обновленные значения в столбце "Value" в датасете data. Результат будет содержать столбец "Value" с числами в типе float.

```python
import re

# Используем регулярное выражение для поиска чисел с плавающей точкой в каждом значении столбца
pattern = r"[-+]?\d*\.\d+|\d+"

# Преобразуем значения столбца в тип float с помощью регулярных выражений и функции float
data['Value'] = data['Value'].apply(lambda x: float(re.findall(pattern, x)[0]))

print("Преобразованный датасет:")
data
```

```python
# Сохраняем новый датасет в CSV-файл
data.to_csv('new_dataset.csv', index=False)
print("Новый датасет сохранен в new_dataset.csv.")
```
