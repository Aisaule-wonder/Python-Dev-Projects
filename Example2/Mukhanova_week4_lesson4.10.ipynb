### Import Pandas

```python
import pandas as pd
```

### загружаем датафрейм

```python
#загружаем датафреймы
orders09 = pd.read_excel("OrdersSept.xlsx")
orders10 = pd.read_excel("OrdersOct.xlsx")
```

```python
orders09.head()
```

```python
orders10.head()
```

```python
#.concat()
orders = pd.concat(objs = [orders09,orders10])
```

```python
orders
```

```python
#.value_counts()
orders.index.value_counts()
```

```python
#чтобы различать источник, можно добавить как key
orders = pd.concat(objs = [orders09,orders10], keys = ["sep","oct"])
```

```python
orders
```

```python
#.value_counts()
orders.index.value_counts()
```

```python
#.loc сентябрь
orders.loc[("sep")]
```

```python
#loc сентябрь, индекс 0
orders.loc[("sep",0)]
```

```python
#loc сентябрь, индекс 0, CustomerID
orders.loc[("sep",0), "CustomerID"]
```

```python
# параметр ingore_index=True чтобы починить индексы
orders = pd.concat(objs = [orders09,orders10], ignore_index = True)
```

```python
orders
```

```python
orders.count()
```

```python
orders.index.value_counts()
```

### `.merge()`

```python
#загружаем датафрейм
customers = pd.read_excel("Customers.xlsx")
customers.head()
```

```python
#inner join
orders.merge(customers, how ="inner", on = "CustomerID").count()
```

```python
#left join
orders.merge(customers, how ="left", on = "CustomerID").count()
```

```python
#right join
orders.merge(customers, how ="right", on = "CustomerID").count()
```

```python
#outer join
orders.merge(customers, how ="outer", on = "CustomerID").count()
```
