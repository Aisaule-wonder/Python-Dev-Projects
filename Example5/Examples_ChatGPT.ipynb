```python
def categorize_rate(rates):
    if not rates:
        return None

    # Рассчитываем среднее значение
    average_rate = sum(rates) / len(rates)

    # Создаем два списка для выше и ниже среднего
    above_average = []
    below_average = []

    # Разделяем значения на категории
    for rate in rates:
        if rate > average_rate:
            above_average.append(rate)
        else:
            below_average.append(rate)

    return above_average, below_average

# Пример использования
rates = [50, 60, 70, 40, 80, 30]
above_average, below_average = categorize_rate(rates)

print("Значения выше среднего:", above_average)
print("Значения ниже среднего:", below_average)
```

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

# Задаем элементы для множеств A, B и C
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
C = {5, 6, 7, 8}

# Создаем Venn-диаграмму
venn_diagram = venn3([A, B, C], set_labels=('A', 'B', 'C'))

# Добавляем метки для областей пересечения
venn_diagram.get_label_by_id('100').set_text(A - (B.union(C)))
venn_diagram.get_label_by_id('010').set_text(B - (A.union(C)))
venn_diagram.get_label_by_id('001').set_text(C - (A.union(B)))
venn_diagram.get_label_by_id('110').set_text(A.intersection(B) - C)
venn_diagram.get_label_by_id('101').set_text(A.intersection(C) - B)
venn_diagram.get_label_by_id('011').set_text(B.intersection(C) - A)
venn_diagram.get_label_by_id('111').set_text(A.intersection(B).intersection(C))

# Отображаем диаграмму
plt.show()
```
