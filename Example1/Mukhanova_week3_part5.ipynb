###  ключевое слово in (keyword)

```python
# IN не только для pandas, но и в целом в Python работает
5 in [5, 4, 3, 3, 2]
```

```python
# при работе с pandas series, IN по умолчанию "смотрит" на index
print(colors)
print("green in colors ", 'green' in colors)
print("3 in colors ", 3 in colors)

#чтобы поиск был по значениям, нужно использовать атрибут values 
'green' in colors.values
```

###  слайсинг по индексу

```python
#индекс в series, как и в листах, начинается с нуля
states = pd.read_excel('states.xlsx', usecols = ['State']).squeeze()
print("states[0]", states[0])
print("states[1]", states[1])
states
```

```python
#индекс в series, как и в листах - end is exclusive
print("states[1:3] ", states[1:3])
```

```python
#отрицательный идекс будет начинать с конца, где [-1:] даст последний элемент
print("last state: states[-1:] ", states[-1:])
print("last state: states[-5:] ", states[-5:])
print("between 10 and 5: states[-10:-5]", states[-10:-5])
```

###  слайсинг по нескольким индексам сразу

```python
# для выбора нескольких индексов, расположенных не по порядку, нужны двойные квадратные скобки
states[[1,3,5,7,9]]
```
