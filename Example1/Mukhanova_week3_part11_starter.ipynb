### Import Pandas

```python
import pandas as pd
```

### загружаем датафрейм

```python
#загружаем данные, students performance
st_df = pd.read_excel('studentsperformance7.xlsx')
st_df
```

```python
#метод .count() чтобы понять, сколько значений в каждой колонке
st_df.count()
```

```python
#метод .info() чтобы понять, есть ли у нас nulls
st_df.info()
```

###  `.dropna()` чтобы удалить NULL 

```python
#метод .dropna() по умолчанию удалит все строки которые содержат хотя бы один NULL, 
#параметр inplace=True чтобы удалить "на самом деле", тем самым изменив исходный датасет
st_df.dropna().info()
```

```python
#метод .dropna() параметр subset чтобы указать конкретные колонки
columns = ["gender", "race/ethnicity", "parental level of education"]
st_df.dropna(subset=columns).info()
```

```python
#метод .dropna() параметр how чтобы удалить строки где все NULL
st_df.dropna(how="all").info()
```

```python
#метод .dropna() параметр axis чтобы указать что удаляем: колонки или строки?
# axis = 1 или axis = 'columns' для колонок
# axis = 0 или axis = 'rows' для строк (по умолчанию удаляет строки)
st_df.dropna(axis = 1, how="all")
```

###  `.fillna()` чтобы заменить NULL 

```python
#загружаем данные, students performance
st_df = pd.read_excel('studentsperformance7.xlsx')
st_df.tail(3)
```

```python
#метод .fillna() чтобы заменить null на что-либо
st_df.fillna("abvgd").tail(3)
```

```python
#метод .fillna() чтобы заменить null на что-либо в конкретной колонке
st_df["gender"].fillna("non-binary").tail(3)
```

```python
#метод .value_counts() параметр dropna по умолчанию True и НЕ показывает кол-во Null
st_df["gender"].value_counts()
```

```python
#метод .value_counts() имеет параметр dropna=False, чтобы увидеть кол-во Null
st_df["gender"].value_counts(dropna=False)
```

### `.isnull()` и `.notnull()`

```python
# isnull() возвращает True если null, False если not null
st_df["gender"].isnull()
```

```python
# отобразить только те строки, где колонка gender is null
missing = st_df["gender"].isnull()
st_df[missing]
```

```python
# notnull() возвращает False если null, True если not null
st_df["gender"].notnull()
```

```python
# отобразить только те строки, где колонка gender НЕ является null
not_missing = st_df["gender"].notnull()
st_df[not_missing]
```
